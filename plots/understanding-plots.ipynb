```python
array1=np.array([[2,3,4],[5,6,8]])
array2=np.array([[1,5,6],[4,7,8]])
plt.plot(array1,array2)
plt.show()
```

```python
x_points=np.array([1,2,4])
y_points=np.array([7,5,1])
plt.plot(x_points,y_points,'s')# i want to mark the poins as o for circle or x or s for square d for diamond
plt.show()
```

```python
#on axis 
ypoints=np.array([1,4,8,9,2,3])
plt.plot(ypoints)
plt.show()
```

```python
ypoints=np.array([1,4,8,9,2,3])
plt.plot(ypoints,'d:r')# d diamond points and r for red
plt.show()
```
